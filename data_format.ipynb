```python
def write_file(path='C:/Users\Crow/Desktop/human_rawdata/uniprot-human-filtered-reviewed%3Ayes.fasta',
               path1="C:/Users/Crow/Desktop/human_rawdata/human_large.txt",
               path2='C:/Users/Crow/Desktop/human_rawdata/human_large_pdata2.txt',
               path3='C:/Users/Crow/Desktop/human_rawdata/out.fasta',
               path4='C:/Users/Crow/Desktop/human_rawdata/human_large_data_clean_P.fasta'):
    
    fr=open(path, 'r')
    fw=open(path3, 'w')
    seq={}
    for line in fr:
        if line.startswith('>'):    #判断字符串是否以‘>开始’
            name=line.split()[0]    #以空格为分隔符，并取序列为0的项。
            seq[name]=''
        else:
            seq[name]+=line.replace('\n', '')
    fr.close()                           

    for i in seq.keys():
        fw.write(i)
        fw.write('\n')
        fw.write(seq[i])
        fw.write('\n')
    fw.close()
    print('fasta file to one line over.')
    
    
    pdata = pd.read_table(path1)
    pdata2 = pdata.groupby(['Protein accession'])['Position'].apply(list)
    print('data.groupby over.')
    # step1 file create 
    fw=open(path2, 'w')
    for i in range(0,len(pdata2)):
        fw.write(pdata2.index[i])
        fw.write('    ')
        #fw.write(str(pdata2[i]))
        for j in pdata2[i]:
            fw.write(str(j))
            fw.write('    ')
        fw.write('\n')
    fw.close()
    print('step1 file over.')
    
    fr = open(path3,'r')
    fw = open(path4, 'w')
    for line in fr:
        for i in range(0,len(pdata2)):
            if line.startswith('>'):
                name = line.split()[0]
            if pdata2.index[i] in name:
                if line.startswith('>'):
                    fw.write('>' + pdata2.index[i])
                    fw.write('    ')
                    for j in pdata2[i]:
                        fw.write(str(j))
                        fw.write('    ')
                    fw.write('\n')
                else:
                    fw.write(line)
    
    fr.close()
    fw.close()
    print('fasta file over')
```

```python
fr=open('C:/Users/Crow/Desktop/human_rawdata/human_small_data_clean_P.fasta', 'r')
fw=open('C:/Users\Crow/Desktop/human_rawdata/data_window.fasta', 'w')
seq={}

for line in fr:
    if line.startswith('>'):
        list = []
        name=line.replace('>','').split()[0]
        seq[name]=''
        pname=line.split()[1:] 
        for i in pname:
            seq[name]+= i + " "
    
fr.close()                           
fw.close()
```

```python
data = pd.DataFrame(columns=['protein','site'])
i = 0
for key in seq.keys():
    for site in seq[key].split():
        
        data.loc[i] = key,site
        
        i += 1
print(data)
```

```python
# 正样本文件输出
# protein   seq   is_train   label
# 正样本滑动窗口文本输出
# window_size
window_size = 51
ws = int((window_size - 1)/2)
path1 = "C:/Users/Crow/Desktop/human_rawdata/human_small_data_clean_P.fasta"
path2 = 'C:/Users/Crow/Desktop/human_rawdata/' + str(window_size) + '_positive.fasta'

fr = open(path1,'r')
fw = open(path2, 'w')
for i in range(0,len(data)):
    fr = open(path1,'r')
    for line in fr:
        if line.startswith('>'):
            name = line.split()[0]
        if data['protein'][i] in name:
            if line.startswith('>'):
                name1 = (name.replace('>','') + ' ' + data['site'][i] + ' 1')
                #fw.write('\n')
            else:
                site = int(data['site'][i])
                seq_raw = line.replace('\n', '')
                lens = len(seq_raw)
                if site-1-ws>=0 and site+ws<=lens:
                    seq = seq_raw[site-ws-1:site+ws]
                if site-1-ws<0 and site+ws<=lens:
                    seq = seq_raw[0:site+ws]
                    for t in range(0,(ws-site)+1):
                        seq = '-' + seq
                if site-1-ws>=0 and site+ws>lens:
                    seq = seq_raw[site-ws-1:]
                    for t in range(0,ws-(lens-site)):
                            seq = seq +  '-'
                if site-1-ws<0 and site+ws>lens:
                    seq = seq_raw[:]
                    for t in range(0,(ws-site)+1):
                        seq = '-' + seq
                    #  检查是否缺失一位
                    for d in range(0,ws-(lens-site)):
                        seq = seq + '-'
                
                fw.write(seq)
                fw.write(' ')
                fw.write(name1)
                fw.write('\n')
                print(seq)


fr.close()
```
