```python
# 原始数据按照site合并
pdata = pd.read_table(path1)

sid = ''
seq = {}
for i in range(0,len(pdata)):
    if(i>0 and pdata['Protein accession'][i] == pdata['Protein accession'][i-1]):
        sid += str(pdata['Position'][i-1]) + " " + str(pdata['Position'][i]) 
        #print('1')
        continue
    else:
        sid += str(pdata['Position'][i]) + " "
        seq[pdata['Protein accession'][i]] = sid
        sid = ''
    
    
    
    

pdata2 = pdata.groupby(['Protein accession'])['Position'].apply(list)
```

```python
print(pdata2[0:10])

# 原始数据输出文本
fr = open(path3,'r')
fw = open('C:/Users/Crow/Desktop/tabacum_data.fasta', 'w')

for line in fr:
    for i in range(0,len(pdata2)):
        if line.startswith('>'):
            name = line.split()[0]
        if pdata2.index[i] in name:
            print(i)
            if line.startswith('>'):
                fw.write('>' + pdata2.index[i])
                fw.write('    ')
                for j in pdata2[i]:
                    fw.write(str(j))
                    fw.write('    ')
                fw.write('\n')
            else:
                fw.write(line)
  
```

```python
fw = open('C:/Users/Crow/Desktop/rice_data/rice_data_11.12.fasta', 'w')

import ssl
 
ssl._create_default_https_context = ssl._create_unverified_context


headers = {'User-Agent':'Mozilla/5.0 (Windows NT 11.0; Win64; x64)'}  


for i in range(0,len(pdata2)):
    name = pdata2.index[i]
    
    #url='https://www.uniprot.org/uniprot/'+ name + '.fasta'
    #s = request.urlopen(url).read().decode('utf8')
    try:
        url='https://www.uniprot.org/uniprot/'+ name + '.fasta'
        req = urllib.request.Request(url=url, headers=headers)  
        s = request.urlopen(req).read().decode('utf8')
        if not(s==''):
            seq = ''
            for q in s.splitlines()[1:]:
                seq = seq + q
        print(seq)
        fw.write('>' + name)
        fw.write('    ')
        for j in pdata2[i]:
            fw.write(str(j))
            fw.write('    ')
        fw.write('\n')
        fw.write(seq)
        fw.write('\n')
    except urllib.error.HTTPError:
        print('HTTPError')
        continue
    except urllib.error.URLError:
        print('URLError')
        continue
    
   
    time.sleep(1)
fw.close()
```
